```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("/content/Heart01.csv")
```

```python
df.info()
```

```python
df1=df.drop(columns=["Timestamp","Patient ID",])
```

```python
df1.info()
```

```python
categorical_cols = df1.select_dtypes(include=['object', 'category']).columns
print("Categorical columns:", categorical_cols)
```

```python
unique_values = df1["Risk Category"].unique()
print(unique_values)
```

```python
unique_values = df1["Gender"].unique()
print(unique_values)
```

```python
df1['Gender'] = df1['Gender'].map({'Male': 1, 'Female': 0})
print(df1)
```

```python
df1['Risk Category'] = df1['Risk Category'].map({'High Risk': 1, 'Low Risk': 0})
print(df1)
```

```python
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
df1=df1.dropna()
```

```python
df1.isnull().sum()
```

```python
duplicates = df1.duplicated()
print(duplicates.sum())
```

```python
cols_to_round = ['Derived_MAP', 'Derived_BMI', 'Derived_HRV','Height (m)', 'Weight (kg)','Body Temperature', 'Oxygen Saturation']
df1[cols_to_round] = df1[cols_to_round].round(2)
print(df1)
df1
```

```python
df1.info()
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = df1.drop('Risk Category', axis=1)
y = df1['Risk Category']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Get feature importance
importance = model.feature_importances_

# Convert to Series for better readability
importances = pd.Series(importance, index=X.columns).sort_values(ascending=False)
print(importances)

# Plot feature importance
importances.plot(kind='bar', figsize=(10,5))
plt.title("Feature Importance from XGBoost")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
print("Train Accuracy:", accuracy_score(y_train, model.predict(X_train)))
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)
print("Cross-validation mean accuracy:", scores.mean())
```
